```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import re
import matplotlib.pyplot as plt
import seaborn as sns
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
from tqdm import tqdm
def clear_text(mbti1):
    lemmatizer=WordNetLemmatizer()
    cleaned_text=[]
    for sentence in tqdm(mbti1.posts):
        sentence=sentence.lower()
        sentence=re.sub('https?://[^\s<>"]+|www\.[^\s<>"]+',' ',sentence)
        sentence=re.sub('[^0-9a-z]',' ',sentence)
        cleaned_text.append(sentence)
    return cleaned_text
```

```python
import plotly.express as px
px.pie(mbti1,names='type',title='Personality type')
```

```python
mbti2=mbti1[0:2][:]
print(mbti2)
mbti2=clear_text(mbti2)
mbti2
```

```python
mbti1.posts=clear_text(mbti1)
mbti1
```

```python
X=mbti1['posts']
Y=mbti1['type']
from sklearn.model_selection import StratifiedKFold
skf=StratifiedKFold(n_splits=39,random_state=None)
skf.get_n_splits(X,Y)
for train_index,test_index in skf.split(X,Y):
  print(train_index,test_index)
  X_train,X_test=X.iloc[train_index],X.iloc[test_index]
  Y_train,Y_test=Y.iloc[train_index],Y.iloc[test_index]
```

```python
from sklearn.preprocessing import LabelEncoder
vectorizer=TfidfVectorizer(analyzer='word',stop_words='english',max_features=1000)
vectorizer.fit(X_train)
```

```python
X1_train=vectorizer.transform(X_train).toarray()
X1_test=vectorizer.transform(X_test).toarray()
```

```python
target_encoder=LabelEncoder()
Y1_train=target_encoder.fit_transform(Y_train)
Y1_test=target_encoder.fit_transform(Y_test)
```

```python
from imblearn.over_sampling import SMOTE
smk=SMOTE(random_state=42)
X_train_res,Y_train_res=smk.fit_resample(X1_train,Y1_train)
X_train_res.shape,Y_train_res.shape
```

```python
from sklearn.metrics import classification_report
```

```python
from sklearn.svm import SVC
model_svc=SVC()
model_svc.fit(X_train_res,Y_train_res)
from sklearn.metrics import classification_report
print('train classification report \n ',classification_report(Y_train_res,model_svc.predict(X_train_res),zero_division=1,target_names=target_encoder.inverse_transform([i for i in range(16)])))
print('test classification report \n ',classification_report(Y1_test,model_svc.predict(X1_test),zero_division=1,target_names=target_encoder.inverse_transform([i for i in range(16)])))
```

```python
from sklearn.ensemble import RandomForestClassifier
random=RandomForestClassifier(class_weight='balanced')
random.fit(X_train_res,Y_train_res)
print('train classification report \n ',classification_report(Y_train_res,random.predict(X_train_res),zero_division=1,target_names=target_encoder.inverse_transform([i for i in range(16)])))
print('test classification report \n ',classification_report(Y1_test,random.predict(X1_test),zero_division=1,target_names=target_encoder.inverse_transform([i for i in range(16)])))
```

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression(solver='liblinear',max_iter=1000,multi_class='ovr')
lg.fit(X_train_res,Y_train_res)
print('train classification report \n ',classification_report(Y_train_res,lg.predict(X_train_res),zero_division=1,target_names=target_encoder.inverse_transform([i for i in range(16)])))
print('test classification report \n ',classification_report(Y1_test,lg.predict(X1_test),zero_division=1,target_names=target_encoder.inverse_transform([i for i in range(16)])))
```

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier(max_depth=30)
tree.fit(X_train_res,Y_train_res)
print('train classification report \n ',classification_report(Y_train_res,tree.predict(X_train_res),zero_division=1,target_names=target_encoder.inverse_transform([i for i in range(16)])))
print('test classification report \n ',classification_report(Y1_test,tree.predict(X1_test),zero_division=1,target_names=target_encoder.inverse_transform([i for i in range(16)])))
```

```python
from sklearn.naive_bayes import MultinomialNB
naive=MultinomialNB()
naive.fit(X_train_res,Y_train_res)
print('train classification report \n ',classification_report(Y_train_res,naive.predict(X_train_res),zero_division=1,target_names=target_encoder.inverse_transform([i for i in range(16)])))
print('test classification report \n ',classification_report(Y1_test,naive.predict(X1_test),zero_division=1,target_names=target_encoder.inverse_transform([i for i in range(16)])))
```
